```python
#Step-9
def stats(s):
    s = pd.Series(s)
    Min = s.min()
    Max = s.max()
    Mean = s.mean()
    Dev = s.std()
    res = [Min,Max,Mean,Dev]
    indx = ["Min","Max","Mean","Dev"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = x.apply(stats)
loc_stats
```

**Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!**




---



---



```python
import pandas as pd
us = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = us['Year']
Y = us['Marriages_per_1000']
Z = us['Divorces_per_1000']

plt.figure(figsize=(12, 8))
plt.plot(X, Y, label = 'Marriages per capita')
plt.plot(X, Z, label = 'Divorces per capita')

plt.xlabel("Year", fontsize=20)
plt.ylabel("Rate (per capita)", fontsize=20)
plt.title("Marriages and divorces per capita in the U.S. between 1867 and 2014.", fontsize=20, fontweight='bold')
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

year = us['Year']
M = us['Marriages_per_1000']

plt.figure(figsize=(12, 8))
plt.plot(year, M, color='blue', label='Marriages per capita')
plt.title('Marriages per capita in US between 1867 and 2014', fontsize=18, fontweight='bold')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Marriages per capita', fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

year = us['Year']
D = us['Divorces_per_1000']

plt.figure(figsize=(12, 8))
plt.plot(year, D, color='maroon', label='Divorces per capita')
plt.title('Divorces per capita in US between 1867 and 2014', fontsize=18, fontweight='bold')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Divorces per capita', fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
```

**Question 7**



**Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!**




---



---



```python
import pandas as pd
us = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us
```

```python
x=us.loc[us.Year.isin([1900,1950,2000]),['Year', 'Marriages_per_1000', 'Divorces_per_1000']]
x
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = x['Year']
Y = x['Marriages_per_1000']
Z = x['Divorces_per_1000']

X1 = np.arange(len(X))

plt.figure(figsize=(15, 10))
plt.bar(X1 - 0.2, Y, 0.4, label = 'Marriages per capita', color='red')
plt.bar(X1 + 0.2, Z, 0.4, label = 'Divorces per capita', color='green')

plt.xticks(X1, X, fontsize=20)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Rate (per capita)", fontsize=20)
plt.title("Marriages and divorces per capita in the U.S between 1900, 1950, and 2000.", fontsize=20, fontweight='bold')
plt.legend(fontsize=15)

plt.show()
```

**Question 8**



**Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!**




---



---



```python
import pandas as pd
deadliest = pd.read_csv('actor_kill_counts.csv')
deadliest
```

```python
import matplotlib.pyplot as plt

y = deadliest['Actor']
x = deadliest['Count']
colors = ['#3B1A82','#D53D7F','#0051C6','#23A4C4','#AD1C68','#3A346C','#632A09','#91AF26','#65A1A6','#64370A']
fig, ax=plt.subplots(figsize=(8,4),dpi=80)
bars=plt.barh(y, x, color=colors)
plt.title('Actors vs. kill count', fontsize=16)
plt.ylabel('Actor (names)', fontsize=12)
plt.xlabel('Kill Count', fontsize=12)
for bar in bars:
    plt.text(bar.get_width() + 5, bar.get_y() + bar.get_height()/2, f'{bar.get_width()}', va='center', fontsize=10, color='black')
plt.show()
```

**Create a pie chart showing the fraction of all Roman Emperors that were

Make sure that the pie chart is an even circle, labels the categories, and shows the

---



---



```python
import pandas as pd
emperor = pd.read_csv('roman-emperor-reigns.csv')
emperor
```
